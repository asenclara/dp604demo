```python
data = [
    (0.563723, 0.382947382, 0.58693748, 0.856758734, 0.8723748),
    (0.564564563, 0.31231233242, 0.58693748, 0.856758734, 0.12312342),
    (0.123123132, 0.382947382, 0.5756713, 0.45345345, 0.8723748)
]
columns = ["A","B","C","D","E"]
df = spark.createDataFrame(data, schema=columns)
```
