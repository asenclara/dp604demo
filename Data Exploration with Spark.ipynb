```python
from pyspark.sql.functions import col, count, when
```

```python
#when(col(c).isNull(), c) checks if the value in column ‘c’ is null. 
#If it is, it returns the column name ‘c’; otherwise, it returns null

#count(...) counts the occurrences where the value in column ‘c’ is null.

#alias(c) assigns the resulting count to a new column with the same name as the original column ‘c’.
null_records = df.select([count(when(col(c).isNull(),c)).alias(c) for c in df.columns])
```

```python
null_records.show()
```

```python
df = df.drop("PassengerId", "Name", "Ticket", "Cabin")
```

```python
median_age = df.select("Age").agg({"Age":"median"}).collect()[0][0]
median_age
```

```python
df = df.na.fill(median_age, subset = ["Age"])
```

```python
null_records = df.select([count(when(col(c).isNull(),c)).alias(c) for c in df.columns])
null_records.show()
```

```python
embarked_mode = df.groupBy("Embarked").count().orderBy(col("count").desc()).first()[0]
embarked_mode
```

```python
df = df.na.fill(embarked_mode, subset = ["Embarked"])
```

```python
null_records = df.select([count(when(col(c).isNull(),c)).alias(c) for c in df.columns])
null_records.show()
```
